## 4. Exploratory Data Analysis

### 4.1 Genre

```python
# some movies have multiple genres separated by commas, we will split them into lists

df['Genre'] = df['Genre'].apply(lambda x: x.split(", ") if isinstance(x, str) else x)
```

```python
genres_expanded = df['Genre'].explode()

genre_counts = genres_expanded.value_counts()

df_genre_counts = genre_counts.reset_index()
df_genre_counts.columns = ['Genre', 'Nombre de films']

df_genre_counts
```

```python
plt.figure(figsize=(10,8))

sns.barplot(x='Genre', y='Nombre de films', data=df_genre_counts)

plt.title('Nombre de films par Genre')
plt.xlabel('Genre')
plt.ylabel('Nombre de films')
plt.xticks(rotation=45, ha='right')

plt.show()
```

### 4.2 Date de sortie

```python
# deleting movies with release date in the future
 
df['Date de sortie'] = pd.to_datetime(
    df['Date de sortie'],
    errors='coerce'
)

df = df[df['Date de sortie'] <= pd.Timestamp.today()]

df['Date de sortie'].max()
```

```python
movies_by_years = df['Date de sortie'].dt.year.value_counts().sort_index()
```

```python
# Création des périodes de 5 ans
df['periode_5_ans'] = (df['Date de sortie'].dt.year // 5) * 5

films_par_periode = df['periode_5_ans'].value_counts().sort_index()

plt.figure(figsize=(12,5))
ax = films_par_periode.plot(kind='bar')

# Ajout des labels au-dessus de chaque barre
for i, value in enumerate(films_par_periode.values):
    ax.text(
        i,
        value,
        f"{value}",
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.title("Nombre de films par période de 5 ans")
plt.xlabel("Période")
plt.ylabel("Nombre de films")
plt.show()
```

```python
cdf = movies_by_years.cumsum() / movies_by_years.sum()

plt.figure(figsize=(10,5))
cdf.plot()
plt.axhline(0.8, linestyle='--')
plt.title("Distribution cumulée des films par année")
plt.ylabel("Proportion cumulée")
plt.xlabel("Année")
plt.show()
```

L’analyse de la distribution temporelle met en évidence une forte concentration de films récents. Afin de réduire le bruit statistique et améliorer la stabilité du moteur de recommandation, une pondération temporelle a été appliquée afin de favoriser les contenus récents tout en préservant la diversité et la richesse historique du catalogue :

```python
# add a temporal weight to the movies based on their release date, giving more importance to recent movies  
current_year = df['Date de sortie'].dt.year.max()

df['poids_temporel'] = 1 / (current_year - df['Date de sortie'].dt.year + 1)
```

### 4.3 Note

```python
print("ANALYSE DES NOTES ET POPULARITÉ\n")
print("=" * 60)

# Statistiques descriptives
print("\nNOTES (vote_average):")
print(df['Note'].describe())
print(f"Médiane: {df['Note'].median()}")
print(f"Films sans note (0): {(df['Note'] == 0).sum():,}")

print("\nPOPULARITÉ:")
print(df['Popularité'].describe())
print(f"Médiane: {df['Popularité'].median()}")

# Distribution des notes
print("\nRépartition des notes:")
bins = [0, 3, 5, 6, 7, 8, 10]
labels = ['Très mauvais (0-3)', 'Mauvais (3-5)', 'Moyen (5-6)', 
          'Bon (6-7)', 'Très bon (7-8)', 'Excellent (8-10)']
df['Note_Catégorie'] = pd.cut(df['Note'], bins=bins, labels=labels, include_lowest=True)
print(df['Note_Catégorie'].value_counts().sort_index())

# Corrélation Note vs Popularité
correlation = df[['Note', 'Popularité']].corr()
print(f"\nCorrélation Note/Popularité: {correlation.iloc[0,1]:.3f}")
```
